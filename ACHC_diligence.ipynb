```
import pandas as pd, numpy as np
import requests
import json
import time
import geopy

final_data = []
# Parameters
coordinate_csv = pd.read_csv("ctc_coordinates_cleaned.csv")
coordinates = coordinate_csv['Coordinates']
states = coordinate_csv['State']
addresses = coordinate_csv['Full Address']
print(coordinates)
# coordinates = ['40.619744, -75.471549']
keywords = ['Opioid, Addiction, Treatment centers, Clinic, Outpatient']
radius = '12070'
api_key = 'AIzaSyCL3KonFLohUIQvmR3hucwgZ9P4lWi50mI' #insert your Places API
for i,coordinate in enumerate(coordinates):
  for keyword in keywords:
    url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+coordinate+'&radius='+str(radius)+'&keyword='+str(keyword)+'&key='+str(api_key)
    while True:
      print(url)
      respon = requests.get(url)
      jj = json.loads(respon.text)
      print(jj)
      results = jj['results']
      for result in results:
        init_coordinate = coordinate
        init_address = addresses[i]
        init_state = states[i]
        name = result['name']
        place_id = result ['place_id']
        lat = result['geometry']['location']['lat']
        lng = result['geometry']['location']['lng']
        vicinity = result['vicinity']
        data = [init_coordinate, init_address, init_state, name, place_id, lat, lng, vicinity]
        final_data.append(data)
        print(data)
      time.sleep(3)
      if 'next_page_token' not in jj:
        break
      else:
        next_page_token = jj['next_page_token']
        url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?key='+str(api_key)+'&pagetoken='+str(next_page_token)
labels = ['Initial Coordinate', 'Initial Address', 'Initial State','Place Name','Place ID', 'Latitude', 'Longitude','Vicinity']
export_dataframe_1_medium = pd.DataFrame.from_records(final_data, columns=labels)
export_dataframe_1_medium.to_csv('export_dataframe_1_medium.csv')
```
